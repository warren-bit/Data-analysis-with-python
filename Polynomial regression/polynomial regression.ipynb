```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
# dividing the dataset in two components x and y
X = df.iloc[:, 1:2].values
y = df.iloc[:, 2:3].values
```

```python
X
```

```python
y
```

```python
# fitting liniear regression to the dataset
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
y_hat = lr.predict(X)
print(y_hat)
```

```python
# fitting polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)

x_poly = poly.fit_transform(X)

poly.fit(x_poly, y)
lr2 = LinearRegression()
lr2.fit(x_poly, y)
y_hat2 = lr2.predict(poly.fit_transform(X))
print(y_hat2)
```

visualizing linear regression results

```python
# regression plot
width  = 10
height = 8
plt.figure(figsize=(width,height))
plt.title("Linear Regression")
sns.regplot(x = "Temperature", y = "Pressure", data = df)
```

```python
# plotting a distrinution plot of actual vs predicted values
plt.figure(figsize=(width,height))
ax1 = sns.distplot(x = y, color = "r", hist=False, label = "Actual values")
sns.distplot(x = y_hat, color = "b", hist=False, label = "Fitted values", ax = ax1)
plt.title("Actual vs Fitted values of Pressure")
plt.xlabel("Pressure")
```

Visualizing polynomial regression results

```python
# regression plot
plt.figure(figsize = (width,height))
plt.scatter(X,y, color = "blue")

plt.plot(X, y_hat2, color = "red")
plt.xlabel("Temperature")
plt.ylabel("Pressure")
plt.title("Polynomial Regression")
plt.show()
```

```python
# plotting a distrinution plot of actual vs predicted values
plt.figure(figsize=(width,height))
ax1 = sns.distplot(x = y, color = "r", hist=False, label = "Actual values")
sns.distplot(x = y_hat2, color = "b", hist=False, label = "Fitted values", ax = ax1)
plt.title("Actual vs Fitted values of Pressure")
plt.xlabel("Pressure")
```
